```python
#Topic : OOPS
#Problem 1: Bank Account Create a class representing a bank account with attributes like account number, 
#account holder name,and balance. Implement methods to deposit and withdraw money from the account.

class bank:
     def __init__(self,name,account_number,balance):
        self.name = name
        self.acc_no = account_number
        self.balance = balance
        self.transactions = 0
        self.passbook_details = []
 
     def withdraw(self,withdraw_amount):
        
        if self.balance >= withdraw_amount:
            self.transactions += 1
            self.balance = self.balance - withdraw_amount
            print(f"\nhi {self.name} {withdraw_amount} Rs. withdraw from your account.\nAvailable balance is {self.balance}..")
            
            self.transaction = {"transaction_no" :self.transactions,"credited_amount" : 0,"debited_amount":withdraw_amount,"Available_balance":self.balance}
            self.passbook_details.append(self.transaction)
        else:
            print(f"\nInsufficient balance, You can withdraw up to {self.balance}..")
        
    
     def diposite(self,diposite_amount):
        if diposite_amount <= 0:
            print("\nplease diposite a valid amount....")
            
        else:
            self.transactions += 1
            self.balance += diposite_amount
            print(f"\nhi {self.name} Available balance is {self.balance}")
            
            self.transaction = {"transaction_no" :self.transactions,"credited_amount" : diposite_amount,"debited_amount":0,"Available_balance":self.balance}
            self.passbook_details.append(self.transaction)
    
     def account_info(self):
        print("\n:::Account Details:::\n=====================")    
        print(f"Account_holder = {self.name}")
        print(f"Account_Number = {self.acc_no}")
        print("\nTransactions Details:")
        for i in range(len(self.passbook_details)):
            for val, value in self.passbook_details[i].items():
                print(f"{val} = {value}",end=" ")
            print()    
        


sudh = bank("sagar",3456333,10000)
sudh.withdraw(2000)
sudh.diposite(123)

sudh.withdraw(20000)
sudh.diposite(0)

sudh.account_info()


```

```python
#Topic : OOPS
#Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle,
#return a vehicle, and display available vehicles.

class Vehicle:
    
    def __init__(self):
        self.vehicles = {1001 : "gt650 bike",1002 : "Activa mopetbike",1003 : "XUV suvcar",1004: "ferrari supercar",1005:"Rickshaw auto"}
        self.on_rent = {}
        
    def rent_vehicle(self,vehicle_id):
        if vehicle_id in self.vehicles and vehicle_id not in self.on_rent :
            self.vehicle_type = self.vehicles.pop(vehicle_id)
            self.on_rent[vehicle_id] = self.vehicle_type
            print(f"you have take a {self.vehicle_type} on rent...")
            print("Available vehicles")
            for idn,vehicle in self.vehicles.items():
                print(f"{idn} = {vehicle}")
        else:
            print(f"There is no vehicle associated with id {vehicle_id}..")
        
    def return_vehicle(self,vehicle_id,vehicle_type):
        if vehicle_id not in self.vehicles and vehicle_id in self.on_rent :
            self.vehicles[vehicle_id] = vehicle_type
            self.on_rent.pop(vehicle_id)
            print(f"you have return a {self.vehicle_type}  which you had taken it on rent...")
            print("Available vehicles")
            for idn,vehicle in self.vehicles.items():
                print(f"{idn} = {vehicle}")
        else:
            print(f"There is no vehicle associated with id {vehicle_id}..")
        
a = Vehicle()
a.rent_vehicle(1001)

a.return_vehicle(1001,"gt650")
```

```python
#Topic : OOPS
#Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add 
#books to the library, borrow books, and display available books.

class Book:
    
    def __init__(self,book_name,author_name):
        self.book_name = book_name
        self.author_name = author_name
        self.is_available = True
        
        
    def borrow(self):
        if self.is_available:
            self.is_available = False
            return True
        else:
            return False
        
    def return_book(self):
        if not self.is_available:
            self.is_available = True
            

class Library:
    def __init__(self):
        self.books = []
        
    def add_book(self,book):
        self.books.append(book)
        
    def display_books(self):
        self.available_books = [book for book in self.books if book.is_available]
        if self.available_books:
            print("Available Books :")
            for book in self.available_books:
                print(f"Book_name :{book.book_name} by Author : {book.author_name}.")
        else:
            print("NO books available in Library..")
            
            
def main():
    
    book_1 = Book("book1","author1")
    book_2 = Book("book2","author2")
    book_3 = Book("book3","author3")
    
    
    library = Library()
    library.add_book(book_1)
    library.add_book(book_2)
    library.add_book(book_3)
    
    book_to_borrow = library.books[1]
    borrowed = book_to_borrow.borrow()
    if borrowed:
        print(f"{library.books[1].book_name} has been borrowed.")
    else:
        print(f"Cannot be borrowed.")
        
    library.display_books()    
        
    Book.return_book(library.books[1])
    print(f"{library.books[1].book_name} has returned.")
    
    library.display_books()



if __name__ == "__main__":
    main()


```

```python
#Topic : OOPS
#Problem 5: Product Inventory Create classes representing a product and an inventory system. 
#Implement methods to add products to the inventory, update product quantity, and display available products.

class Product:
    
    def __init__(self,product_id,product_name,product_price,product_qty):
        self.product_id = product_id
        self.product_name = product_name
        self.product_price = product_price
        self.product_qty = product_qty
    
        
    def __str__(self):
        return f"Product ID: {self.product_id}, Name: {self.product_name}, Price: ${self.product_price}, Quantity: {self.product_qty}"
    
        
        
class Inventory:
    def __init__(self):
        self.products = []
        
    def add_product(self,product): 
        self.products.append(product)
        
    def update_quantity(self,product_id,quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.product_qty = quantity
                return True
        return False
    
    def display_products(self):
        print("\nAvailable Products:")
        self.available_products = [product for product in self.products if product.product_qty > 0]
        if self.available_products:
            for product in self.available_products:
                print(f"Product_Id: {product.product_id}.  Name: {product.product_name}.  Price: {product.product_price}.  Quantity: {product.product_qty}")
        else:
            print(f"Zero Products")
        
        
def main():
    inventory = Inventory()
    
    product_1 = Product(1,"watch",350,5)
    product_2 = Product(2,"shoes",1500,4)
    product_3 = Product(3,"wallet",500,8)
    
    
    inventory.add_product(product_1)
    inventory.add_product(product_2)
    inventory.add_product(product_3)
    
    inventory.display_products()
    
    if inventory.update_quantity(2,12):
        print(f"\nUpdated product2 quantity")
    else:
        print(f"Product Quantity can't Updated")
    
    inventory.display_products()
    
    
if __name__ == "__main__":
    main()

```

```python
#Topic : OOPS
#Problem 6: Shape Calculation Create a class representing a shape with attributes like length,
#width, and height. Implement methods to calculate the area and perimeter of the shape.

class Shape:
    
    def __init__(self,length, width, height):
        self.length = length
        self.width = width
        self.height = height
        
        
    def calc_area(self):
        #Formulae 2lw + 2lh + 2wh
        self.area = 2*(self.length*self.width + self.length*self.height + self.width*self.height)
        return self.area
    
    def calc_perimeter(self):
        #Formulae 4(L+W+H)
        self.perimeter = 4 * (self.length + self.width + self.height)
        return self.perimeter
    
    
def main():
    shape_1 = Shape(34,5,23)
    
    
    shape_1_area = shape_1.calc_area()
    shape_1_perimeter = shape_1.calc_perimeter()
    
    print(f"Length = {shape_1.length} Width = {shape_1.width} Height = {shape_1.height}\nArea = {shape_1_area}\nPerimeter = {shape_1_perimeter}")
    
        
if __name__ == "__main__":
    main()
        



```

```python
#Topic : OOPS
#Problem 7: Student Management Create a class representing a student with attributes like student ID,
#name, and grades. Implement methods to calculate the average grade and display student details.

class Student:
    def __init__(self,student_id,name):
        self.student_id = student_id
        self.name = name
        self.grades = []
        
    def add_grade(self,grade):
        self.grades.append(grade)
        
    def calc_avg_grade(self):
        if self.grades:
            return sum(self.grades)/len(self.grades)
        else:
            return 0
    def display_details(self):
        print("\nStudent Details:")
        print(f"Student_Id = {self.student_id}")
        print(f"Student_name = {self.name}")
        print(f"Student_Grades = {', '.join(map(str,self.grades))}")
        print(f"Average_grade = {self.calc_avg_grade():.2f}")
        
        
def main():
    student1 = Student(101, "John Smith")
    student1.add_grade(85)
    student1.add_grade(92)
    student1.add_grade(78)

    student2 = Student(102, "Jane Doe")
    student2.add_grade(90)
    student2.add_grade(88)
    student2.add_grade(95)

    student1.display_details()
    print()
    student2.display_details()

    
if __name__ == "__main__":
    main()

```

```python
#Topic : OOPS
#Problem 9: Social Media Profile Create a class representing a social media profile with attributes 
#like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

class Social_media:
    
    def __init__(self,username,post_name,keyword):
        self.username = username
        self.posts = [{post_name:[keyword]}]
        print(f"{username} your 1st Post has been Uploded on Social Media...")
    
    #Adds a Post
    def add_post(self,username,post_name,keyword):
        if self.username == username:
            posts_counter = len(self.posts)
            self.posts.append({post_name:[keyword]})
            print(f"\nAdding Post {posts_counter+1}:\n{self.username} your {posts_counter+1}.{post_name} has been Uploaded on Social Media...")
        else:
            print(f"\nInvalid Username: Username doesnt match...")
    
    #Add keyword to specific post
    def add_keyword(self,username,post_name,new_keyword):
        if self.username == username:
            for post in self.posts:
                if post_name in post:
                    post[post_name].append(new_keyword)
                    print(f"\nAdding Keyword to Post:\nHi {username} the keyword '{new_keyword}' has been added to the post {post_name}...")
                    break
            else:
                print(f"\nHi {username} there is no post matching with '{post_name}'...")

        else:
            print(f"\nInvalid Username: Username doesnt match...")
    
    #Finding all post which tags with specific keyword 
    def search_post_by_keyword(self,username,keyword):
        if self.username == username:
            found_posts = []
            for post in self.posts:
                for post_name , keywords_lst in post.items():
                    if keyword in keywords_lst:
                        found_posts.append((post_name,keywords_lst))
            if found_posts:
                counter = 1
                print(f"\nFinding Posts by Keyword:\n{username} your Matched Posts with keyword '{keyword}'.:")
                for post_name,keywords_lst in found_posts:
                    print(f"{counter}.{post_name}\n  Keywords:{', '.join(keywords_lst)}") 
                    counter += 1
            else:
                print("\nNo posts found with the specified keyword")
                
        else:               
            print(f"\nInvalid Username: Username doesnt match...")
            
    #Displays total posts
    def display_posts(self,username):
        if self.username == username:
            counter = 1
            print(f"\nDisplaying Total Posts:\nUsername '{username}' Posts:")
            for post in self.posts:
                for post_name,keywords in post.items():
                    print(f"{counter}.{post_name}:\n  Keywords:{', '.join(keywords)}")
                    counter+=1
                    
        else:
            print(f"\nInvalid Username: Username doesnt match...")
    
    
    #Delete the specific Posts
    def delete_post(self, username, post_name):
        if self.username == username:
            delete_post = None  
            for post in self.posts:
                for postname, keywords in post.items():
                    if postname == post_name:
                        delete_post = post  
                        break
            if delete_post:
                self.posts.remove(delete_post)
                print(f"\nDeleting Post:\n{self.username} you deleted Post '{post_name}'")
                print(f"Remaining Posts after Deletion :",end=" ")
                self.display_posts(username)
            else:
                print(f"\nNo posts found with name {post_name}.")
        else:
            print(f"\nInvalid Username: Username doesn't match...")
        
            
            
def main():
    
    #Creating instance of class and adding 1st Post.
    ashish = Social_media("Ashish","RiverRafting","Adventure")
    
    #Adding keywords to 1st Post
    ashish.add_keyword("Ashish","RiverRafting","Swimmer")
    ashish.add_keyword("Ashish","RiverRafting","Photography")
    
    #Adding 2nd Post
    ashish.add_post("Ashish","WeddingVibes","SamWedsManu")
    
    #Adding keywords to 2nd Post
    ashish.add_keyword("Ashish","WeddingVibes","JanWeddingShoot")
    ashish.add_keyword("Ashish","WeddingVibes","Photography")
    
    #Adding 3rd post
    ashish.add_post("Ashish","Treaking","Adventure")
    
    #Adding 4th post
    ashish.add_post("Ashish","Waterpark","bathing")
    #Adding Keyword
    ashish.add_keyword("Ashish","Waterpark","waterSprinklers")
    
    #Searching post by posts keyword
    ashish.search_post_by_keyword("Ashish","Photography")
    
    #Display Posts
    ashish.display_posts("Ashish")
    
    #Delete Posts
    ashish.delete_post("Ashish","Waterpark")
    
    #Display Posts
    ashish.display_posts("Ashish")
    
if __name__ == "__main__":
    main()
    
    
```

```python
##Topic : OOPS
#Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates.
#Implement methods to add tasks, mark tasks as completed, and display pending tasks

class TodoList:
    
    def __init__(self):
        self.tasks = []
        
    def add_task(self,task_name,due_date):
        self.task = {"Name" : task_name, "Due_date" : due_date, "Completed" : False}
        self.tasks.append(self.task)
        print(f"Task '{task_name}' has been added in TodoList.")
        
    def mark_task_as_completed(self,task_name):
        for task in self.tasks:
            if task["Name"] == task_name:
                task["Completed"] = True
                print(f"Task - '{task_name}' has been marked as Completed.. ")
                break
                
    def display_pending_tasks(self):
        pending_tasks = [task for task in self.tasks if not task["Completed"]]
        if pending_tasks:
            print("Pending Tasks:")
            for task in pending_tasks:
                print(f"- {task['Name']} (Due on: {task['Due_date']})")
        else:
            print("No pending tasks.")

def main():

    todo_list = TodoList()

    todo_list.add_task("Finish project", "2023-10-15")
    todo_list.add_task("Buy groceries", "2023-10-12")
    todo_list.add_task("Go to the gym", "2023-10-18")

    todo_list.mark_task_as_completed("Buy groceries")

    todo_list.display_pending_tasks()          
                
                
                
if __name__ == "__main__":
    main()
                
                
                
```
